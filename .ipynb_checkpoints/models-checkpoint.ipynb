```python
# Data split
vectorizer = CountVectorizer(ngram_range=(1, 2))
X = vectorizer.fit_transform(df['text'])
y = df["sentiment"]
#le = preprocessing.LabelEncoder()
#y = le.fit_transform(df['sentiment'])


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=1)

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
model = LogisticRegression(
    C=100, 
    multi_class="multinomial")
score_metric="f1_macro"
params = ['max_iter']
param_grid = { params[0]: [700, 1000]}
cv=3
grid_classifier=GridSearchCV(model,param_grid,cv=cv, scoring=score_metric, verbose=3)
grid_classifier.fit(X_train, y_train)

```

```python
grid_classifier.best_params_
```

```python
grid_classifier.get_params
```

```python
import scikitplot as skplt
import matplotlib.pyplot as plt
y_true_merged = y_test# ground truth labels
y_probas_merged = grid_classifier.predict_proba(X_test)# predicted probabilities generated by sklearn classifier
skplt.metrics.plot_roc(y_true_merged, y_probas_merged, figsize=(20,10), cmap="tab20", plot_macro=False)
plt.show()
```

```python
from tensorflow.keras import Sequential, layers, callbacks
model = Sequential([
    layers.Dense(32, activation="relu", input_shape=[X_train.shape[1]]),
    layers.Dropout(0.3),
    layers.BatchNormalization(),
    layers.Dense(16, activation="relu"),
    layers.Dropout(0.3),
    layers.BatchNormalization(),
    layers.Dense(8, activation="relu"),
    layers.Dropout(0.3),
    layers.BatchNormalization(),
    layers.Dense(1, activation="sigmoid")
])
```

```python
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["categorical_accuracy"]
)
```

```python
early_stopping = callbacks.EarlyStopping(
    patience=10,
    min_delta=0.001,
    restore_best_weights=True,
)
```
